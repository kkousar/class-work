1D plot for sin and cos functions

```python
import matplotlib.pyplot as plt
import numpy as np
import math as m

x=np.linspace(0,2*m.pi,100);
y=np.sin(x);
q=np.cos(x);
r=np.sin(x**2);

plt.plot(x,y, 'b-', label='y=sin(x)')
plt.plot(x,q, 'c-', label='y=cos(x)')
#plt.plot(x,r, 'm-', label='y=sin(x^2)')
plt.legend(loc='upper left');

plt.show();
```

plotting 2d

```python
from matplotlib import RcParams
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib.cm as cm
from pylab import rcParams
rcParams['figure.figsize']=5,3

#vector plot
x=np.linspace(2,-2,10);
y=np.linspace(2,-2,10);

xv,yv=np.meshgrid(x,y);

plt.figure()
plt.quiver(xv,yv,-xv**3,-yv**3)

plt.savefig('vector.pdf');
plt.show();
```

#3D plotting

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm



plt.gca(projection='3d');
#vector plot
x=np.linspace(-10,10,5);
y=np.linspace(-10,10,5);
z=np.linspace(-10,10,5)

xv,yv,zv=np.meshgrid(x,y,z);

plt.quiver(xv,yv,zv,-xv,-yv,-zv)
plt.show();
```
