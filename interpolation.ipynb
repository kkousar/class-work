#interpolation
Thus interpolation is the process of finding the value of function for any intermediate value of the independent variable

#linear method
quicklatex.com-1fdaccf93d8bcbf6128f0a169911ba1a_l3.svg

```python
def inter_linear(d,x):
  output = d[0][1] + (x - d[0][0]) * ((d[1][1] - d[0][1])/(d[1][0] - d[0][0]))
  return output



data=[[1,10.23],[3,23.20]];
x=2.5;
print(inter_linear(data,x));
```
